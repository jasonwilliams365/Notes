**Part 2: Short Answer Questions**

1. Briefly explain why understanding the history of the internet can help a backend developer today.
 ANSWER: It helps backend developer understand how-why the Internet works the way it does the help make coding more streamlined.

2. Describe the role of DNS in a web request.
ANSWER: DNS allows users to enter a domain name such as Facebook.com or US.gov rather than having to enter the entire IP address in order to visit a website

3. What is the main difference between HTTP and HTTPS?
ANSWER: HTTPS is the secure version of HTTP.

4. Name two common HTTP methods and explain when you would use each.
ANSWER: GET and PUT. GET is used to fetch data from server without modding. PUT fully replaces existing data/existing resource.

5. How can the OSI model help developers troubleshoot network-related issues?
ANSWER: Allows developers to understand the structure of a computer and a computer network for more pinpointed troubleshooting.

6. Why is Docker important in modern backend development workflows?
ANSWER: One reason would be that Docker helps to minimize the 'works on my machine' issues.

7. Give an example of a situation where Kubernetes would be beneficial for your application.
ANSWER: Kubernetes would be useful for a big project that needs constant monitoring or patching. It would allow developers to push small patches/updates to keep applications at near-peak performance.

8. How does AWS simplify infrastructure management for backend developers?
ANSWER: AWS provides many services that allow backend developers to organize and de-complicate tasks such as scaling or creating a VR environment/server.

9. What is the significance of using proper HTTP status codes in API responses?
ANSWER: Status codes help the user to know what the issue is they are facing and what may need to be done in order to correct those issues.

10. Explain one core principle of DevOps culture and why it matters in backend development.
ANSWER: One core principle of DevOps culture is to never feel you HAVE to memorize everything. DevOps is a broad topic/category that encompasses a lot of different terms and rules. The ability to know how to research and troubleshoot is very important in DevOps.
