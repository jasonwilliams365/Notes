### Part 2: Web Fundamentals

**2.1 HTTP and HTTPS Protocols**  
- HTTP definition and stateless nature:  HTTP = Hypertext Transport Protocol
Stateless nature: this means every request is independent(data is not stored unless cookies or data files are kept)
- HTTPS and security (TLS/SSL):  HTTPS = Hypertext Transport Protocol Secure
- Why HTTPS matters for modern backends:  Keeps webpages and info secure vs. HTTP where it is available to anyone.

**2.2 HTTP Methods (Verbs)**  
- GET:  Retrieve - used to fetch data from server without modding
- POST:  Create - Used to send data to server  to create resource
- PUT:  Replace - Fully replaces existing data/existing resource
- DELETE:  Remove: Used to delete a resource from server
- PATCH:  Partially Replace - Partially mods data/resource
- Application in RESTful API design:  
Micro-services: Carrys OOP apps to create one complete service

**2.3 The Request-Response Cycle**  
- Client request, server processing, server response:  
- Role of load balancers and servers:  Distributes various requests to multiple users

**2.4 HTTP Status Codes**  
- 2xx (Success): 200 OK
- 3xx (Redirection): 301 Moved Permanently (redirect)  
- 4xx (Client errors): Common codes (404, 403, 400, 401)  - auth error
- 5xx (Server errors): 500, 503, etc. - Internal Server Error 
- Importance of proper status codes in API design: Helps with troubleshooting
